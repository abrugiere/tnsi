## Instructions conditionnelles et boucles

### Instructions conditionnelles

- La structure la plus simple est le `if` :
```
if condition :
	instructions
```
Le mot **condition**  d√©signe une expression bool√©enne (qui vaut soit `True` soit `False`) et le mot **instructions** d√©signe une instruction ou un bloc d‚Äôinstructions √©crites sur plusieurs lignes √† effectuer lorsque la condition est vraie. Exemple :

``` python
if n % 2 == 0 :
	print("Le nombre est pair")
```


- La structure `if-else` ajoute une alternative pour ex√©cuter une autre instruction ou bloc d‚Äôinstructions √† effectuer lorsque la condition est fausse :
```
if condition :
	instructions1
else :
	instructions2 
```

Par exemple :
``` python
if n % 2 == 0 :
	print("Le nombre est pair")
else :
	print("Le nombre est impair")
```


- La structure `if-elif-else` ajoute plusieurs alternatives conditionnelles :
```
if condition1 :
	instructions1
elif condition2 :
	instructions2
elif condition3 :
	instructions3
...
else :
	instructions4
```

Par exemple :
``` python
if n % 4 == 0 :
	n = n // 4
elif n % 4 == 1 :
	n = ( 3 * n + 1 ) // 4
elif n % 4 == 2 :
	n = n // 2
else :
	n = ( 3 * n + 1 ) // 2

```


> ‚ö†Ô∏è **Remarques :**  
> - Les mots `if` et `elif` sont toujours suivis par une expression qui prendra la valeur `True` ou `False`, la ligne se termine par le symbole `:` (deux points)
>- le mot `else` est toujours imm√©diatement suivi par le symbole `:` (deux points)
>- C‚Äôest **l‚Äôindentation** qui permet de d√©limiter les blocs d‚Äôinstructions √† ex√©cuter lorsque les conditions sont v√©rifi√©es
>- Les mots `True`, `False` et `None` sont les rares mots en Python dont l‚Äô√©criture commence par une lettre majuscule



### Boucles conditionnelles

La structure est la suivante :
```
while condition :
	instructions
```

La structure est identique √† celle du `if`. Le bloc d‚Äôinstructions indent√© qui suit est ex√©cut√© tant que la condition est v√©rifi√©e. C‚Äôest ce bloc d‚Äôinstructions qui doit modifier la valeur de la condition pour qu‚Äôelle ne soit plus v√©rifi√©e et que le programme sorte du bloc et continue son ex√©cution.
``` python
while a  >= b :
    a = a - b
    n = n + 1
```

### Boucles non conditionnelles (it√©ratives)

Une boucle non conditionnelle sera r√©p√©t√©e `n` fois, `n` √©tant connu √† l‚Äôavance, la structure est la suivante :
```
for i in range(n) :
	instructions
```

Ce code est √©quivalent √† :
```
i = 0
while i <= n :
	instructions
	i = i + 1
```

Dans les deux cas, une variable `i` est cr√©√©e et prendra successivement les valeurs `0`,`1`,`2`‚Ä¶,`n-1` pour la boucle for et `0`,`1`,`2`,...,`n` pour la boucle while et les instructions seront ex√©cut√©es `n` fois.


La boucle `for` en Python propose d‚Äôautres possibilit√©s, si on consid√®re objet comme une s√©quence de plusieurs √©l√©ments (chaine de caract√®res, liste‚Ä¶) on pourra utiliser :
```
for element in objet :
	instructions
```

Exemple :

````python
for car in 'Bonjour' :
    print(3 * car)
````

> ‚ö†Ô∏è **Remarques :**
> - `for i in range(f)` : `i` prendra la valeur des entiers successifs de `0` inclus √† `f` **exclu**
> - `for i in range(d,f)` : `i` prendra la valeur des entiers successifs de `d` inclus √† `f` **exclu**
> - `for i in range(d,f,p)` : `i` prendra la valeur des de `d` inclus √† `f` **exclu avec un pas de** `p`.

## Les fonctions

### D√©finition

Une fonction se d√©finit de la mani√®re suivante :
```
def nomDeLaFonction(argument1,argument2):
    """ aide sur la fonction : docstring
    (facultatif) """
    Corps De La Fonction
```

`def` est un mot cl√© du langage Python, les arguments sont s√©par√©s de virgules.
Le corps de la fonction est un bloc de code indent√©.
‚ÄÉ
- Si le corps de la fonction contient l'instruction `return`, alors l'appel de la fonction est une expression qui a donc une valeur.
- S'il n'y a pas d'instruction `return` dans le corps de la fonction, alors l'appel de la fonction renverra la valeur `None`. Ce type de fonction s'appelle une **proc√©dure**.

Exemple :

``` python
def volume(longueur,largeur,hauteur):
    """ renvoie le volume d'un parall√©l√©pip√®de
    dont on fournit les trois dimensions en arguments."""
    
    return longueur * largeur * hauteur
```








### Espace et port√©e des variables
**L‚Äôespace local** d‚Äôune fonction contient les param√®tres qui lui sont pass√©s, et les variables d√©finies au sein de celle-ci.  
Pour qu‚Äôune variable soit utilis√©e dans une fonction, il faut que la variable appartienne √† son espace local ou √† l‚Äôespace qui appelle la fonction.  
Une fonction ne peut pas modifier, par affectation, la valeur d'une variable ext√©rieure √† son espace local. 

````python
x=3

def f(x):
    x+=2
    print(x)
````

````python
f(x)
````

````python
print(x)
````

La variable `x` utilis√©e dans la fonction est distincte de la variable `x` d√©finie au d√©but du programme `x=3` et n'existe plus apr√®s l'appel de la fonction. 

Apr√®s l'instruction `f(x)`, l'espace local de la fonction `f` est d√©truit, C'est pour cel√† que `print(x)` renvoie `3` et non `5`.


> üí° Il existe un moyen de modifier avec une fonction des variables ext√©rieures √† celle-ci. On utilise pour cela des variables **globales** mais ce proc√©d√© est √† √©viter car il nuit √† la portabilit√© d‚Äôune fonction.
