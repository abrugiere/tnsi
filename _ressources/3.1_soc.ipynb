# Composants intégrés d’un système sur puce

##  Loi de Moore et miniaturisation progressive
### La Loi de Moore

En 1965, Gordon Moore postule que le nombre de transistors présents sur une puce de microprocesseur doublera tous les deux ans.  
Cette prédiction s'est révélée étonnamment juste (à quelques approximations près) et les équipements électroniques n'ont depuis jamais cessé de devenir toujours plus performants / miniatures / économes en énergie.

### Évolution de la taille et de la puissance des ordinateurs

Le premier ordinateur : l'**ENIAC** (**E**lectronic **N**umerical **I**ntegrator **a**nd **C**omputer) achevé en 1945 pesait plus de 30 tonnes et occupait 167 m2. Elle fonctionnait à 100 kHz : pouvait effectuer 100 000 opérations par seconde.  



L'invention du **transistor** en 1947 puis des ordinateurs à **circuit intégré** à partir de 1958 vont permettre la miniaturisation des ordinateurs. C'est à partir de cette date que l'utilisation de l'informatique a explosé.  


Dans les années 70 les **ordinateurs personnels** font leur apparition, les microprocesseurs de cette époque, tels que l’Intel 4004, fonctionnaient à environ 740 kHz (740 000 opérations par seconde).  

Dans les années 1990, la puissance de calcul a considérablement augmenté. Les processeurs de cette époque fonctionnaient généralement à des vitesses de l’ordre de quelques **mégahertz (MHz)**, soit des millions d’opérations par seconde ouvrant la porte à des applications graphiques plus avancées dans les jeux vidéo notamment.  

Les processeurs des années 2000 ont franchi la barre du **gigahertz (GHz)**, atteignant des vitesses d’environ 3 GHz, soit environ 3 milliards d’opérations par seconde. Cette puissance de calcul accrue a permis le développement de logiciels de retouche photo et de montage vidéo grand public.  

À l’ère moderne, les processeurs pour ordinateurs personnels dépassent régulièrement les 4 GHz, les GPU (graphic processing unit), utilisés pour le rendu graphique et l’intelligence artificielle, sont devenus des acteurs majeurs de la puissance de calcul.  


## Composition d'un ordinateur actuel

Dans un ordinateur, chaque composant a un rôle spécifique. Les composants communiquent entre eux par des **bus** de différentes vitesses. Chaque composant est remplaçable, et il est possible d'ajouter de nouveaux composants sur la carte mère qui possède des slots d'extension.  
Les composants les plus notables sont : 
- le microprocesseur (CPU)
- la mémoire vive (ou RAM : Random Access Memory)
- la carte graphique avec son processeur intégré (GPU)
- des ports d'extension  : USB, HDMI, VGA…
- des composants de communication : carte Wifi, Bluetooth, 4G…

## Tout un pc sur une seule puce : les SoC

### Principe
Le principe d'un système sur puce ou **S**ystem **o**n a **C**hip (*SoC*) est d'intégrer au sein d'une puce unique un ensemble de composants habituellement physiquement dissociés dans un ordinateur classique (ordinateur de bureau ou ordinateur portable).  
On peut retrouver ainsi au sein d'une même puce :
- le microprocesseur (CPU)
- la carte graphique (GPU)
- la mémoire RAM
+ éventuellement des composants de communication (WiFi, Bluetooth…)

### Exemples
Les SoC sont présents depuis maintenant plusieurs années dans les smartphones (puces Snapdragon ou encore Apple A5) et les micro-ordinateurs **RaspberryPi** ou **Arduino** et consoles de ju (Nintendo Switch, Playstation Vita)  
  

  

Depuis 2020 avec le processeur M1, Apple utilise un SoC (architecture ARM) dans ses ordinateurs MacBook et a déployé cette architecture sur tous ses produits.  

Microsoft travaille sur une système d'exploitation supportant cette architecture, et il semblerait que celle-ci devienne un future standard pour tous nos ordinateurs personnels ([clubic](https://www.clubic.com/actualite-523644-windows-11-microsoft-persuadee-que-les-nouveaux-pc-portables-arm-rivaliseront-vraiment-avec-les-macbook-m3.html), [generation-nt](https://www.generation-nt.com/actualites/arm-processeur-ordinateur-part-marche-strategie-2047633)).  


### Pourquoi utiliser ces systèmes sur puces plutôt qu'une architecture classique ?

L'engouement pour ces systèmes est lié au fait que ceux-ci apportent de multiples avantages face aux systèmes classiques :
- L'**efficacité énergétique**. Les SoC sont conçus pour consommer moins d'énergie que leurs homologues, ce qui les rend idéaux pour les appareils mobiles ou embarqués alimentés par des batteries.
- Moindre **encombrement** permettant l'utilisation en embarqué ou de réduire la taille des ordinateurs.
- Chauffe moins dissipation passive de la chaleur, pas besoin de **refroidissement**.
- Moindre **coût** (forte automisation du processus, gros volumes de production)

Cependant, les SoC souffrent de quelques inconvénients :
- Impossibilité de **choisir** indépendamment ses composants
- Remplacement ou **ajout** d'un composant impossible
- La **panne** d'un seul composant entraîne la panne totale du SoC
