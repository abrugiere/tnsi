# Composants intégrés d’un système sur puce

---
## Évolution de la taille et de la puissance des ordinateurs

Le premier ordinateur : l'**ENIAC** (**E**lectronic **N**umerical **I**ntegrator **a**nd **C**omputer) achevé en 1945 pesait plus de 30 tonnes et occupait 167 m2. Il pouvait effectuer 100 000 opérations par seconde (fréquence de calcul = 100 kHz).  


L'invention du **transistor** en 1947 puis des ordinateurs à **circuit intégré** à partir de 1958 vont permettre la miniaturisation des ordinateurs. C'est à partir de cette date que l'utilisation de l'informatique a explosé.  


Dans les années 70 les **ordinateurs personnels** font leur apparition, les microprocesseurs de cette époque, tels que l’_Intel 4004_, fonctionnaient à environ 740 kHz (740 000 opérations par seconde).  

Dans les années 1990, la puissance de calcul a considérablement augmenté. Les processeurs de cette époque fonctionnaient généralement à des vitesses de l’ordre de quelques **mégahertz (MHz)**, soit des millions d’opérations par seconde ouvrant la porte à des applications graphiques plus avancées dans les jeux vidéo notamment.  

Les processeurs des années 2000 ont franchi la barre du **gigahertz (GHz)**, atteignant des vitesses d’environ 3 GHz, soit environ 3 milliards d’opérations par seconde. Cette puissance de calcul accrue a permis le développement de logiciels de retouche photo et de montage vidéo grand public.  

À l’ère moderne, les processeurs pour ordinateurs personnels dépassent régulièrement les 4 GHz, les GPU (Graphic Processing Unit) utilisés pour le rendu graphique et l’intelligence artificielle, sont devenus des acteurs majeurs de la puissance de calcul.

## Composition d'un ordinateur

Dans un ordinateur, chaque composant a un rôle spécifique. Les composants communiquent entre eux par des **bus** de différentes vitesses. Chaque composant est remplaçable, et il est possible d'ajouter de nouveaux composants sur la carte mère qui possède des slots d'extension.  
Les composants les plus notables sont : 
- le microprocesseur (CPU)
- la mémoire vive (ou RAM : Random Access Memory)
- la carte graphique avec son processeur intégré (GPU)
- des ports d'extension  : USB, HDMI, VGA…
- des composants de communication : carte Wifi, Bluetooth, 4G…  
  

[Cette vidéo](https://www.youtube.com/watch?v=vJch4_Yjbl8) vous détaillera la liste des ces composants.

## Tout un pc sur une seule puce : les SoC

On entend souvent dire que les téléphones portables (smartphones) sont de véritables ordinateurs, ce qui est vrai. Comment peut-on retrouver dans un smartphone les mêmes composants que dans un PC : CPU, RAM, carte graphique et interfaces réseau ?  
La solution est de placer tous ces composants dans une puce unique d'une dizaine de cm<sup>2</sup>  

Le principe d'un **système sur puce** ou **SoC** (**S**ystem **o**n a **C**hip) est d'intégrer au sein d'une puce unique un ensemble de composants habituellement physiquement dissociés dans un ordinateur classique (ordinateur de bureau ou ordinateur portable).  
On peut retrouver ainsi au sein d'une même puce :  
- le microprocesseur (CPU)
- la carte graphique (GPU)
- la mémoire RAM  
+ éventuellement des composants de communication (WiFi, Bluetooth…)  

Les SoC sont présents depuis maintenant plusieurs années dans les smartphones (puces Snapdragon ou encore Apple A5) et les micro-ordinateurs **RaspberryPi** ou **Arduino** et consoles de jeu (Nintendo Switch, Playstation Vita)  
  

  

Depuis 2020 avec le processeur M1, Apple utilise un SoC (architecture ARM) dans ses ordinateurs MacBook et a déployé cette architecture sur tous ses produits.  

En 2024, Microsoft travaille sur son système d'exploitation pour supporter cette architecture et rattraper son retard sur Apple. Il semblerait que le SoC devienne un futur standard pour tous nos ordinateurs personnels ([clubic](https://www.clubic.com/actualite-523644-windows-11-microsoft-persuadee-que-les-nouveaux-pc-portables-arm-rivaliseront-vraiment-avec-les-macbook-m3.html), [generation-nt](https://www.generation-nt.com/actualites/arm-processeur-ordinateur-part-marche-strategie-2047633)).

## Pourquoi utiliser un SoC plutôt qu'une architecture classique ?

L'engouement pour ces systèmes est lié au fait que ceux-ci apportent de multiples avantages face aux systèmes classiques :
- L'**efficacité énergétique**. Les SoC sont conçus pour consommer moins d'énergie que leurs homologues, ce qui les rend idéaux pour les appareils mobiles ou embarqués alimentés par des batteries.
- Un **moindre encombrement** permettant l'utilisation en embarqué ou de réduire la taille des ordinateurs.
- Ils sont aussi **moins sujets à la surchauffe**. Ils dispersent moins de chaleur, ce qui les rend plus robustes car ils ne requièrent pas de composants de gestion de chaleur.
-  Les composants de l'architecture sont plus petits par rapport à ceux des puces ordinaires. Ce qui amoindrit le **coût par puce**. (forte automisation du processus, gros volumes de production)

Cependant, les SoC souffrent de quelques inconvénients :
- Impossibilité de **choisir** indépendamment ses composants
- **Changement** d'un composant impossible (pour ajout ou modification)
- La **panne** d'un seul composant entraîne la panne totale du SoC
- Pour le moment, ils sont encore **incompatibles** avec de nombreux logiciels et le développement de leur compatibilité est complexe et nécessite de nombreuses modifications.
