> - Ecrivez ci-dessous une fonction `cesar(decalage,phrase)` qui renvoie la cha√Æne `phrase` pass√©e en param√®tre d√©cal√©e √† l'aide du code de C√©sar d√©cal√©e de `decalage`  
> üí° Vous allez parcourir la chaine de caract√®res lettre par lettre, ensuite vous trouverez le code unicode de chaque lettre (avec `ord`) et lui ajouterez le d√©calage. Ensuite il vous suffit de retraduire le nouveau code en caract√®re (avec `chr`).  
> **ATTENTION** : si vous d√©calez de 5 la lettre Z vous allez obtenir le code 95 soit le caract√®re "\_", il va donc falloir retrancher 26 √† votre code si celui-ci d√©passe 90. Vous prendrez √©galement en compte le fait de ne pas remplacer le carac√®re si c'est un espace.

```python
# √† compl√©ter
def cesar(d,p):
    res=''
    for l in p :
        if l == ' ' :
            res = res + l
        else:
            new=ord(l)+d
            if new >90:
                new-=26
            res = res + chr(new)
    return res
```

```python
# V√©rification : doit renvoyer 'DQDM TI VAQ'
cesar(8,'VIVE LA NSI')
```

> - Ecrivez maintenant une fonction `rasec(decalage,phrase)` qui renvoie la cha√Æne `phrase` pass√©e en param√®tre d√©cod√©e √† l'aide du code de C√©sar  

```python
# √† compl√©ter
def rasec(d,p):
    res=''
    for l in p :
        if l == ' ' :
            res = res + l
        else:
            new=ord(l)-d
            if new <65:
                new+=26
            res = res + chr(new)
    return res
```

```python
# V√©rification
rasec(17,'SIRMF MFLJ RMVQ UVTYZWWIV CV TFUV')
```

> - Ecrivez ci-dessous une fonction `subs(phrase)` qui renvoie la phrase `phrase` pass√©e en param√®tre chiffr√©e par substitution √† l'aide du dictionnaire ci-dessus.  
> üí° En utilisant un dictionnaire, l'impl√©mentation devient tr√®s simple

```python
# √† compl√©ter
def subs(p):
    res=''
    dic={
        "A":"Z",
        "B":"Y",
        "C":"X",
        "D":"W",
        "E":"V",
        "F":"U",
        "G":"T",
        "H":"S",
        "I":"R",
        "J":"Q",
        "K":"P",
        "L":"O",
        "M":"N",
        "N":"M",
        "O":"L",
        "P":"K",
        "Q":"J",
        "R":"I",
        "S":"H",
        "T":"G",
        "U":"F",
        "V":"E",
        "W":"D",
        "X":"C",
        "Y":"B",
        "Z":"A",
        " ":" "
    }
    
    for l in p :
        res = res + dic[l]
    return res
```

```python
# V√©rification : doit renvoyer 'EREV OZ MHR'
subs('VIVE LA NSI')
```
