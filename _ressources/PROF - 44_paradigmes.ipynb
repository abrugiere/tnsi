### Dfinition
Le paradigme fonctionnel cherche  viter au maximum les effets de bord, dit autrement, en programmation fonctionnelle on va viter de modifier les valeurs des variables. Pour se faire, on va chercher au maximum  utiliser les fonctions (d'o le nom de programmation fonctionnelle), mais ces fonctions ne devront pas modifier les variables. Les fonctions utilises en programmation fonctionnelle sont parfois appeles **"fonctions pures"** : le rsultat retourn par une fonction pure doit uniquement dpendre des paramtres passs  la fonction et pas des valeurs externes  la fonction (elle ne doit pas non plus engendrer d'effet de bord)

Ci-dessous, notre fonction revue en *"fonction pure"* :

````python
#  excuter
def fct(i):
    if i > 5:
        return True
    else :
        return False
````

````python
#  excuter
fct(5)
````

- **Rcursion** : Sans affectation ni boucle, la programmation fonctionnelle passera souvent par des fonctions rcursives.

####  Exercice 4 : La fonction filter

>- Ecrivez ci-dessous une fonction `filtre` qui prend deux paramtres : une fonction boolenne (qui renvoie `True` ou `False` ) et une liste. Elle renvoie une nouvelle liste contenant les lments filtrs  l'aide de la fonction boolenne.  
>Par exemple : 
>``` Python
>    def pair(n):
>        return n%2==0    
>    
>    filtre(pair,[1,2,3,4])

````python
#  complter
def filtre(f,li):
    newli=[]
    for elt in li:
        if f(elt):
            newli.append(elt)
    return newli
````

````python
# Vrification
def pair(n):
        return n%2==0

filtre(pair,[1,2,3,4])
````

>-  La fonction `filter` existe sous python. Elle s'utilise presque comme celle que vous venez d'crire, mais elle renvoie un **objet** qu'il faudra convertir en liste avec la fonction `list()` pour l'utiliser :

````python
#  excuter
l = filter(pair,[1,2,3,4])
list(l)
````

####  Exercice 5 : La fonction reduce

>- Ecrivez ci-dessous une fonction `reduire` qui prend deux paramtres : une fonction et une liste.
>Au premier appel, les deux premiers lments de la liste sont passs en paramtres. Ensuite,
>le rsultat de cet appel et llment suivant de la liste sont passs en paramtres, et ainsi de suite...  
>
>Exemple : En supposant que la fonction `somme(x,y)` renvoie la somme des lments x et y ,  
>alors  
>`reduit(somme,[a,b,c,d])` renverra `somme(somme(somme(a,b),c),d)` soit `(((a+b)+c)+d)`

````python
#  complter
def reduire(f,li):
    valeur = f(li[0],li[1])    
    for i in range(2,len(li)):
        valeur = f(valeur,li[i])
    return valeur
````

````python
# Vrification
def somme(x,y):
    return x+y

l=reduire(somme,[1,2,3,4])
print(l)
````

>- Dcrivez l'excution de ces dernires lignes

>-  La fonction `reduce` existe sous python. Elle s'utilise presque comme celle que vous venez d'crire, il faut pour l'utiliser importer le module `functools` :

````python
#  excuter
from functools import reduce
l=reduce(somme,[1,2,3,4])
print(l)
````
