## Le paradigme impératif

La programmation impérative est le paradigme de programmation le plus ancien. Il définit un programme comme une **séquence** clairement définie d’instructions informatiques.

La programmation impérative repose sur des notions qui vous sont familières :
- la séquence d'instructions (les instructions d'un programme s'exécutent l'une après l'autre)
- l'affectation (on attribue une valeur à une variable, par exemple : `a = 5`)
- l'instruction conditionnelle (`if` / `else`)
- la boucle (`while` et `for`)

Le paradigme de programmation impérative est très concret et fonctionne en étroite collaboration avec le système. Le code est aisément compréhensible, mais l’écriture de nombreuses lignes de code source est nécessaire pour arriver à des programmes complexes.

## La programmation fonctionnelle


### Un exemple

Analysons tout d'abord la fonction suivante :


````python
# à exécuter
def fct():
    if i > 5:
        return True
    else :
        return False
````

Maintenant, affectons une valeur à `i` et exécutons la fonction

````python
# à exécuter
i = 5
fct()
````

Enfin, affectons une nouvelle valeur à `i` et réexécutons la fonction

````python
# à exécuter
i = 22
fct()
````

Comme vous pouvez le constater, sans changer le contenu de la fonction ni son appel, nous avons pu lui faire retourner deux valeurs différentes.  

Cette fonction est dite à [effet de bord](https://fr.wikipedia.org/wiki/Effet_de_bord_(informatique)) car elle utilise une variable extérieure à la fonction (`i`)  


### Définition
Le paradigme fonctionnel cherche à éviter au maximum les effets de bord, dit autrement, en programmation fonctionnelle on va éviter de modifier les valeurs des variables. Pour se faire, on va chercher au maximum à utiliser les fonctions (d'où le nom de programmation fonctionnelle), mais ces fonctions ne devront pas modifier les variables. Les fonctions utilisées en programmation fonctionnelle sont parfois appelées **"fonctions pures"** : le résultat retourné par une fonction pure doit uniquement dépendre des paramètres passés à la fonction et pas des valeurs externes à la fonction (elle ne doit pas non plus engendrer d'effet de bord)

Ci-dessous, notre fonction revue en *"fonction pure"* :

````python
# à exécuter
def fct(i):
    if i > 5:
        return True
    else :
        return False
````

````python
# à exécuter
fct(5)
````

````python
# à exécuter
fct(22)
````

````python
# à exécuter
fct(5)
````

#### 💻 Exercice 1 : Fonction pure
>La fonction suivante ne respecte pas le paradigme fonctionnel puisqu'elle modifie la valeur d'une variable extérieure à la fonction :

````python
# à exécuter
def ajout(i):
    l.append(i)
````

````python
# à exécuter
l = [4,7,3]
ajout(8)
print(l)
````

>- Proposez ci-dessous une version de la fonction ajout qui respecte le paradigme fonctionnel

````python
# à compléter
def ajout2(li,i):
    li2=[]
    for e in li:
        li2.append(e)
    li2.append(i)
    return li2
````

````python
# Vérification
l = [4,7,3]
print(ajout2(l,8))
print(l)
````

### Caractéristiques de la programmation fonctionnelle

- **Pas "Comment ?" mais "Quoi ?"** : En programmation fonctionnelle, on décrira le rapport qui existe entre les données et le résultat que l'on veut obtenir plutôt que la séquence des traitements.
Par exemple en **Haskell** le code suivant renverra la somme des 100 premiers entiers naturels pairs tels que le carré de la somme d'un de ces nombres additionné à 2 est divisible par 137 :
``` Haskell
    sum $ take 100 $ filter (\x -> (x + 2)^2 `mod` 137 == 0) [0,2,..]
```

- La **transparence référentielle** : En programmation fonctionnelle, il n'y a pas d'affectation. Lorsqu'on écrit `a = 1`, a peut être remplacé par 1 tout au long du programme.

- **Idempotence** : On obtient toujours le même résultat pour chaque appel d'une fonction. Il faut donc éviter au maximum les variables globales.

- **Concurrence** et parallélisation : Grâce à l'idempotence et la transparence référencielle, on peut facilement se libérer de la contrainte de chronologie.

- **Pas d'affectation : des liens** : Conséquence de la transparence référentielle, une expression étant liée à une valeur, si on veut changer cette valeur, il faut créer une nouvelle expression. Par exemple, au lieu de :
``` Python
    nom = "James"
    nom = nom + "Bond"
```
on préferera :
``` Python
    prenom = "James"
    nom = "Bond"
    appellation = prenom + nom
```



- **Récursion** : Sans affectation ni boucle, la programmation fonctionnelle passera souvent par des fonctions récursives. 

- **Les fonctions sont des objets comme les autres** : En référence au chapitre sur la calculabilité (à venir), les fonctions pourront prendre d'autres fonctions en argument.
En Python, on pourra utiliser l'opérateur `lambda` qui permettra de créer des **fonctions anonymes** selon la syntaxe :
``` Python
    lambda arguments: image
```
Par exemple on créera *"à la volée"* la fonction $x  -> 2x + 1$ :
``` Python
    lambda x: 2*x+1
```
Qu'on peut lire _"Une fonction qui prend en paramètre `x`  et renvoie `2x + 1`"_

#### 💻 Exercice 3 : La fonction map

>- Ecrivez ci-dessous une fonction `applique` qui prend deux paramètres : une fonction et une liste. Elle renvoie une nouvelle liste contenant les images par la fonction des éléments de la liste initiale.  
>Par exemple : 
>``` Python
>    def carre(x):
>        return x ** 2
>    
>    applique(carre,[1,2,3,4])
>    >>> [1,4,9,16]
>```


````python
# à compléter
def applique(f,li):
    newli=[]
    for elt in li :
        newli.append(f(elt))
    return newli
````

````python
# Vérification 1
def carre(x):
        return x ** 2

applique(carre,[1,2,3,4])
````

````python
# Vérification 2
applique(len,['Joe','Jack','William','Averell'])
````

>- Que renvoie ce dernier appel ?

>- 💡 La fonction `map` existe sous python. Elle s'utilise presque comme celle que vous venez d'écrire, mais elle renvoie un **objet** qu'il faudra convertir en liste avec la fonction `list()` pour l'utiliser :

````python
# à exécuter
l = map(carre,[1,2,3,4])
list(l)
````

#### 💻 Exercice 4 : La fonction filter

>- Ecrivez ci-dessous une fonction `filtre` qui prend deux paramètres : une fonction booléenne (qui renvoie `True` ou `False` ) et une liste. Elle renvoie une nouvelle liste contenant les éléments filtrés à l'aide de la fonction booléenne.  
>Par exemple : 
>``` Python
>    def pair(n):
>        return n%2==0    
>    
>    filtre(pair,[1,2,3,4])
>    >>> [2,4]
>```


````python
# à compléter
def filtre(f,li):
    newli=[]
    for elt in li:
        if f(elt):
            newli.append(elt)
    return newli
````

````python
# Vérification
def pair(n):
        return n%2==0

filtre(pair,[1,2,3,4])
````

>- 💡 La fonction `filter` existe sous python. Elle s'utilise presque comme celle que vous venez d'écrire, mais elle renvoie un **objet** qu'il faudra convertir en liste avec la fonction `list()` pour l'utiliser :

````python
# à exécuter
l = filter(pair,[1,2,3,4])
list(l)
````

#### 💻 Exercice 5 : La fonction reduce

>- Ecrivez ci-dessous une fonction `reduire` qui prend deux paramètres : une fonction et une liste.
>Au premier appel, les deux premiers éléments de la liste sont passés en paramètres. Ensuite,
>le résultat de cet appel et l’élément suivant de la liste sont passés en paramètres, et ainsi de suite...  
>
>Exemple : En supposant que la fonction `somme(x,y)` renvoie la somme des éléments x et y ,  
>alors  
>`reduit(somme,[a,b,c,d])` renverra `somme(somme(somme(a,b),c),d)` soit `(((a+b)+c)+d)`

````python
# à compléter
def reduire(f,li):
    valeur = f(li[0],li[1])    
    for i in range(2,len(li)):
        valeur = f(valeur,li[i])
    return valeur
````

````python
# Vérification
def somme(x,y):
    return x+y

l=reduire(somme,[1,2,3,4])
print(l)
````

>- Décrivez l'exécution de ces dernières lignes

>- 💡 La fonction `reduce` existe sous python. Elle s'utilise presque comme celle que vous venez d'écrire, il faut pour l'utiliser importer le module `functools` :

````python
# à exécuter
from functools import reduce
l=reduce(somme,[1,2,3,4])
print(l)
````

## La programmation orientée Objet


La conception objet s'efforce de calquer la représentation informatique sur les entités physiques. Le concept de **réification** consiste donc à partir d'objets réels, par exemple des objets *Catalogue* ou *Voiture*.


Un exemple typique est la voiture.  
Un développeur a besoin d'intégrer l'idée de voiture dans son programme, cela peut être pour un jeu vidéo de course, une simulation du trafic routier ou tout autre programme ayant besoin du concept de voiture.  
Il va utiliser un langage orienté objet pour créer une **classe** « Voiture » ; cette classe va permettre de stocker deux choses :
- **L'état d'une voiture** : le développeur stocke dans l'état des informations sur une voiture, comme sa couleur, sa marque ou son nombre de chevaux. L'état d'un objet est donc composé de variables et peut être amené à évoluer. Ces variables sont appelées des attributs. Un objet peut aussi contenir d'autres objets. Dans le cas présent il pourrait y avoir un tableau de quatre variables contenant des objets « Roue ».  
- **Le comportement** : c'est un ensemble de fonctions (appelées méthodes) propres à toutes les voitures et qui permettent de savoir ce qu'un objet peut faire. Dans le cas de la classe « Voiture », il pourrait y avoir une méthode démarrer, accélérer ou arrêter. Les méthodes ont accès à l'état de l'objet et peuvent donc interagir avec.

- La programmation orientée objet permet de mélanger état et comportement d'un objet.
- Un objet représente un concept, une idée ou un élément du monde physique.
- La classe correspond à la structure générale de l'objet et la définition de ses méthodes, l'objet correspond à une instance de la classe.

## Autres paradigmes


### La programmations déclarative
Lorsque l'on programme, il est parfois nécessaire d'obtenir un résultat précis, indépendamment de l'état actuel du programme ou de la machine. C'est à dire que l'on souhaite que la machine réalise une tâche, sans avoir besoin de décrire les différentes opérations nécessaires pour y arriver.  
Les langages de programmation déclaratifs permettent de décrire précisément le résultat que l'on souhaite, tout en laissant la gestion du "comment nous voulons l'obtenir" à l'ordinateur.

L'exemple typique de programmation déclarative est le **SQL** (**S**tructured **Q**uery **L**anguage)

``` SQL
    CREATE TABLE users (
        id INTEGER,
        name VARCHAR(50),
        PRIMARY KEY(id)
    );

    INSERT INTO users (id, name) VALUES (1, 'Luke');

    INSERT INTO users (id, name) VALUES (2, 'Anakin');

    SELECT id, name FROM users;

```

Ce code permet de déclarer à la machine que l'on souhaite créer, insérer puis récupérer les champs `id` et `name` de la table `users`, sans pour autant lui décrire comment y arriver. Le développeur ici ne s'intéresse pas au contexte ou à l'état du programme qui enverra le résultat.  

Nous étudierons les bases du langage SQL durant l'année de terminale


### La programmation événementielle 
Il est courant qu'un programme interagisse avec l'extérieur en échangeant des données en entrée et en sortie. Cependant il peut aussi échanger des événements, et déclencher des actions en fonctions d'autres actions extérieures. C'est le cas des sites Web dynamiques, qui réagissent pour changer l'état ou l'apparence de la page, par exemple lorsque l'on clique sur un bouton. La programmation événementielle se focalise ainsi sur l'exécution de code en fonction dʼévénements, plutôt qu'une simple exécution séquentielle des instructions.

- *Python* permet de faire de la programmation événementielle avec notament le module `Tkinter`.
- *Javascript* et *Node.js* sont les langages de la programmation événementielle du Web.
- *Vert.x* est le langage événementiel pour Java.
- *Visual Basic* est le langage événementiel célèbre de Microsoft.

## 💡 Utiliser le bon paradigme

>- Quel que soit le paradigme choisi et le mode d'écriture d'un programme, il sera traduit sous forme de code machine **impératif** : donner à la machine des instructions élémentaires pour lui dire comment effectuer ses aller/retours entre le processeur et la mémoire.
>
>- Il est important de bien comprendre qu'un programmeur doit maitriser **plusieurs paradigmes** de programmation (impératif, objet ou encore événementiel). En effet, il sera plus facile d'utiliser le paradigme objet dans certains cas alors que dans d'autres situations, l'utilisation du paradigme fonctionnelle sera préférable. Être capable de Choisir le "bon" paradigme en fonction des situations fait partie du bagage de tout bon programmeur.
>
>- La tendance actuelle est de ne pas attacher le langage à un paradigme mais de permettre l'adoption de **plusieurs paradigmes au sein d'un même langage** afin de s'adapter à différentes situations. Python permet pa exemple d'utiliser les paradigmes impératifs, objet ou encore événementiel (avec par exemple le module Tkinter).
