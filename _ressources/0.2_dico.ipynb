## GÃ©nÃ©ration des dictionnaires

> Comme pour les listes, les dictionnaires acceptent la **construction par comprÃ©hension**

````python
d = {x:x**2 for x in range (1,5)}
````

````python
d
````

Revient Ã  Ã©crire en une seule ligne le code suivant :

````python
d={}
for x in range(1,5):
    d[x]=x**2
````

````python
d
````

Voici une version plus complexe utilisÃ©e pour coder un message par un dÃ©calage _(code de CÃ©sar)_.

````python
cesar = {chr(65+i): chr(65+(i+7)%26) for i in range (26)}
````

````python
cesar
````

La fonction ```chr``` renvoie le caractÃ¨re correspondant au code donnÃ© en paramÃ¨tre. Les caractÃ¨res 'A','B',...,'Z' sont codÃ©s par les nombres 65,66,...,90.

````python
message = 'BONJOUR'
code = ''
for c in message:
    code = code + cesar[c]

code
````
