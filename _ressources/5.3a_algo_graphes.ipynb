## Introduction

Un parcours de graphe est un algorithme consistant  **explorer tous les sommets en suivant les artes** qui les relient  partir d'un sommet initial. Ces parcours sont notamment utiliss pour rechercher un plus court chemin (et donc dans les GPS) ou pour trouver la sortie d'un labyrinthe    
  
> **Parcourir simplement le dictionnaire ou la matrice d'adjacence dun graphe nest pas considr comme un parcours de graphe.**  


Contrairement  un parcours d'arbre, o l'on part systmatiquement de la racine puis on suit un "ordre" prcis (prfix, suffixe...), il faudra tout d'abord choisir un noeud de dpart puis mmoriser les sommets dja visits ou dcouverts.

## Parcours en largeur ou BFS (Breadth First Search)

### Principe
L'algorithme de parcours en largeur suit le principe suivant :  partir d'un sommet, on atteint **d'abord ses voisins puis les voisins de ses voisins** et ainsi de suite jusqu' visiter tous les sommets.  
Dans le parcours en largeur, on visite tous les sommets autour du sommet de dpart : dabord les sommets  une distance de 1, puis ceux  une distance de 2  On l'appelle galement parcours **concentrique** et il sert notamment  la recherche des plus courts chemins dans un graphe.

> _Notez qu'il n'y a pas une unique faon de parcourir un graphe en largeur l'ordre de parcours des voisins tant arbitraire._

### Exemple avec un graphe non orient
nonori.gif

Dans le graphe ci-dessus, effectuons le parcous en largeur ** partir du sommet A** en choisissant de parcourir les voisins par ordre alphabtique :
- On commence par le sommet **A**
- Ensuite on visite les voisins de A : **B**, **C** et **D**
- Puis les voisins de B : **E** (A a dj t visit)
- Puis les voisins de C : **F** (A et D ont dj ,t visits)
- Puis les voisins de D : **G** (A, C et E ont dj t visits)
- Puis les voisins de E : Pas de nouveau voisin non dj visit
- Puis les voisins de F : Pas de nouveau voisin non dj visit
- Puis les voisins de G : Pas de nouveau voisin non dj visit

Le parcours donne donc **A B C D E F G**, mais il aurait pu donner **A D C B G E F** si nous avions choisi de parcourir les voisins dans l'ordre inverse



### Exemple avec un graphe orient

Avec un graphe orient, le principe est exactement le mme. **On remplace simplement les voisins par les successeurs**. Ceci va naturellement restreindre les possibilits et tous les sommets ne seront pas forcment visits

Dans le graphe ci-dessus, effectuons le parcours en largeur  partir du sommet **A** :

- On commence par le sommet **A** 
- Ensuite on visite les successeurs de A : **B** et **C**
- Puis les successeurs de B : **E**
- Puis les successeurs de C : **F**
- Puis les successeurs de E : Aucun sommet
- Puis les successeurs de F : **G**
- Puis les successeurs de G : **D** (E a dj t visit)
- Puis les succsseurs de D : A, C  et E ont dj t visits

Le parcours donne donc **A B C E F G D**, mais il aurait pu donner **A C B F E G D**

## Parcours en profondeur ou DFS (Depth First Search)

L'algorithme de parcours en profondeur suit le principe suivant :  partir d'un sommet, on va aller **le plus loin possible** sans se proccuper des autres voisins non visits. Il explore ainsi chaque chemin depuis le sommet de dpart jusqu' atteindre un cul-de-sac ou avoir visit tous les sommets, puis revient au dernier sommet o il tait possible de prendre un chemin diffrent. Une application de ce type de parcours est la rsolution de **labyrinthes**.

> _Ici aussi, il n'y a pas une unique faon de parcourir un graphe en profondeur le choix des artes parcourues tant arbitraire._

### Exemple avec un graphe non orient
nonori.png

Dans le graphe ci-dessus, effectuons le parcours en profondeur ** partir du sommet A** :
- On commence par le sommet **A**
- Ensuite on visite le sommet **C**
- Puis le sommet **F**
- Puis le sommet **G**
- Puis le sommet **E**
- Puis le sommet **D**
- Plus de chemin possible, on remonte  E
- Ensuite on visite le sommet **B**

Le parcours donne donc **A C F G E D B**, mais il aurait pu donner **A B E G F C D** ou **A D E G F C B** si nous avions choisi de parcourir les artes diffremment
