## Le Markdown

### C'est quoi ? 
>  Comme le HTML, le markdown est un langage qui permet de mettre en forme du texte en utilisant des __balises__.

### Pourquoi l'utiliser ?
- Il est plus simple pour mettre en forme du texte que *Word* ou un traitement de texte
- On peut l'utiliser **en ligne** et **gratuitement**
- Il peut tre intgr dans les **Notebooks** pour alterner du cours et des exercices

### Exemples :

- Pour mettre un mot en gras, il suffira de l'entourer avec la balise `**` (ou `__`), un mot entour de balises `~~` sera barr, de balises `*` (ou `_`) sera en italique...  
  `Voici un **exemple** d'utilisation du ~~language~~ langage *Markdown*`  
  donnera :  
>  Voici un **exemple** d'utilisation du ~~language~~ langage *Markdown*  

- Les titres seront prcds des balises # (une balise : niveau 1, 2 balises : niveau 2, etc.)  
  ```
  # Titre de niveau 1  
  ## Titre de niveau 2  
  ### Titre de niveau 3
  ```
  donnera :
  # Titre de niveau 1
  ## Titre de niveau 2
  ### Titre de niveau 3
