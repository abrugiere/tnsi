## Parcours

Les arbres √©tant une structure hi√©rarchique, leur utilisation implique la n√©cessit√© d'un parcours des valeurs stock√©es. Par exemple pour toutes les r√©cup√©rer dans un certain ordre, ou bien pour en chercher une en particulier.
Il existe plusieurs mani√®res de parcourir un arbre. (parcourir un arbre = passer par tous les noeuds)  

### Parcours en largeur d'abord (ou BFS : **B**readth **F**irst **S**earch )
Le parcours en largeur d'abord est un parcours √©tage par √©tage (de haut en bas) et de gauche √† droite : On part de la racine, puis on visite son fils gauche puis son fils droit, puis les petits-fils de la racine (toujours de la gauche vers la droite)

par1bfs.jpg
L'ordre des lettres parcourues est donc **T-Y-O-P-H-N**

### Parcours pr√©fixe (ou pr√©ordre)
Le parcours pr√©fixe est un parcours **en profondeur d'abord** (ou DFS : **D**epth **F**irst **S**earch). Chaque noeud est visit√© avant que ses sous-arbres le soient : On part de la racine, puis on visite son sous-arbre gauche (et √©ventuellement le sous-arbre gauche de celui-ci, etc.) avant de remonter et de redescendre vers le sous-arbre droit.

par2pre.jpg
L'ordre des lettres parcourues est donc **T-Y-P-O-H-N**

### Parcours infixe (ou en ordre)
Le parcours infixe est aussi un parcours **en profondeur d'abord** (ou DFS : **D**epth **F**irst **S**earch). Chaque noeud est visit√© apr√®s son sous-arbre gauche mais avant son sous-arbre droit. On part donc de la feuille la plus √† gauche et on remonte par vagues sucessives. Un noeud ne peut pas √™tre visit√© si son sous-arbre gauche ne l'a pas √©t√©.
par3inf.jpg
L'ordre des lettres parcourues est donc **P-Y-T-H-O-N**

### Parcours postfixe (ou postordre)
Le parcours postfixe est aussi un parcours **en profondeur d'abord** (ou DFS : **D**epth **F**irst **S**earch). Chaque noeud est visit√© apr√®s que ses sous-arbres le soient. On part donc de la feuille la plus √† gauche, et on ne remonte √† un noeud p√®re que si ses sous-arbres ont tous √©t√© visit√©s.
par4pos.jpg
L'ordre des lettres parcourues est donc **P-Y-H-N-O-T**

### Pour s'y retrouver
> Dans le cas d'un parcours en largeur d'abord on affiche tous les noeuds situ√©s √† une profondeur n avant de commencer √† afficher les noeuds situ√©s √† une profondeur n+1.
>  
> Pour les parcours en profondeur, les trois mots-cl√©s parlent de la place du p√®re par rapport √† ses fils. Ensuite, il faut toujours se souvenir qu'on traite le fils gauche avant le fils droit.
>- _pr√©_ veut dire **avant** => **pr√©fixe** : le p√®re doit √™tre le premier par rapport √† ses fils.
>- _in_ veut dire **au milieu** => **infixe** : le p√®re doit √™tre entre son fils gauche (trait√© en premier) et son fils droit.
>- _post_ veut dire **apr√®s** => **postfixe** : le p√®re ne doit √™tre trait√© que quand ses deux fils (gauche d'abord, droite ensuite) l'ont √©t√©.
